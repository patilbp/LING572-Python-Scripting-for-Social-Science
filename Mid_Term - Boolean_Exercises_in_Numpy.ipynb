Write a for loop that creates a dictionary containing the counts of all the words

Call your dictonary `wd_cts`.  If your program is correct `wd_cts`
should contain the following count:


````python
# Solution :

wd_counts = dict()
for w in sad_story.split():
    if w in wd_counts:
        wd_counts[w] += 1
    else:
        wd_counts[w] = 1
````

# 2.5

Write a **function** `find_word_forms` that takes a string `text_str` and a word `form` as its **two** arguments and returns all **words** in `text_str` that end with `form`.
Write this without importing any Python modules.

Here are some tests that your function should pass:

```
>>> find_word_forms(sad_story, 'ing')
['bring']
>>> find_word_forms(sad_story, 'ed')
['lived']
>>> find_word_forms("He 's sending birds on the wing .", 'ing')
['sending', 'wing']
>>> find_word_forms(sad_story, 'g')
['long', 'bring']
```

Hint:  You may find it helpful to remember the
string method `endswith`.

````python
# Solution :

def find_word_forms(text_str, form):
  text_str_words = text_str.split()
  result = []
  for word in text_str_words:
      if word.endswith(form):
          result.append(word)
  return result
````

````python
find_word_forms(sad_story, 'ing')
````

````python
find_word_forms(sad_story, 'ed')
````

````python
find_word_forms("He 's sending birds on the wing .", 'ing')
````

````python
find_word_forms(sad_story, 'g')
````

##### **Logic Explanation - (2.5) :**

- I have defined the function **`find_word_forms`** by passing the text string and the word form.
- Next, I have splitted the words of the text using **`split()`**, as it might be present in form of a string or a list.
- Then, I have defined the result list as blank at initial point.
- Further, I have searched for the word in Text String which ends with expected form using the function **`endswith`**.
- Finally, whichever words aho have the form present into them, will get added to the Result list.
